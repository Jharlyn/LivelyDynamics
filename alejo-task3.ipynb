```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
image_size = (500, 500, 3)
background_image = np.zeros(image_size)

background_image[:, :, 0] = np.linspace(0, 1, image_size[0])[:, None]
background_image[:, :, 2] = np.linspace(0.2, 0.9, image_size[1])

w, h, c = image_size
```

```python
def Circle(center, radius, color):
    return mpatches.Circle(center, radius, edgecolor = 'black', facecolor = color)

def Ellipse(center, width, height, edgecolor, facecolor):
    return mpatches.Ellipse(center, width, height, edgecolor = edgecolor, facecolor = facecolor)

def Wedge(center, radius, theta1, theta2, edgecolor, facecolor):
    return mpatches.Wedge(center, radius, theta1, theta2, edgecolor=edgecolor, facecolor=facecolor)
```

```python
donut_bread = Circle((0.53 * w, 0.52 * h), 0.4 * min(w, h), 'beige')
```

```python
ear1 = Wedge((0.25 * w, 0.25 * h), 0.2 * min(w, h), 120, 0, 'black', 'beige')
ear2 = Wedge((0.75 * w, 0.25 * h), 0.2 * min(w, h), 180, 90, 'black', 'beige')
ear3 = Wedge((0.25 * w, 0.25 * h), 0.12 * min(w, h), 120, 0, 'black', 'pink')
ear4 = Wedge((0.75 * w, 0.25 * h), 0.12 * min(w, h), 180, 90, 'black', 'pink')
```

```python
donut_cream = Circle((0.5 * w, 0.5 * h), 0.4 * min(w, h), 'pink')
```

```python
donut1_hole = Ellipse((0.5 * w, 0.5 * h), 2 * 0.2 * 0.4 * min(w, h), 2 * 0.2 * 0.4 * min(w, h), 'black', 'beige')
donut2_hole = Ellipse((0.51 * w, 0.507 * h), 2 * 0.17 * 0.4 * min(w, h), 2 * 0.17 * 0.4 * min(w, h), 'black', 'purple')
```

```python
eye1 = Ellipse((0.35 * w, 0.25 * h), 0.12 * min(w, h), 0.15 * min(w, h), 'black', 'black')
eye2 = Ellipse((0.65 * w, 0.25 * h), 0.12 * min(w, h), 0.15 * min(w, h), 'black', 'black')
```

```python
pupil1 = Circle((0.32 * w, 0.22 * h), 0.015 * min(w, h), 'white')
pupil2 = Circle((0.62 * w, 0.22 * h), 0.015 * min(w, h), 'white')
```

```python
mouth = Wedge((0.5 * w, 0.35 * h), 0.05 * min(w, h), 0, 180, 'black', 'black')
```

```python
patches = [
    donut_bread,
    ear1,
    ear2,
    ear3,
    ear4,
    donut_cream,
    donut1_hole,
    donut2_hole,
    eye1,
    eye2,
    pupil1,
    pupil2,
    mouth,
]

#Sprinkles
rectangles = [
    ((0.4 * w, 0.6 * h), 0.05 * w, 0.02 * h, 'white'),
    ((0.55 * w, 0.7 * h), 0.06 * w, 0.02 * h, 'green'),
    ((0.35 * w, 0.7 * h), 0.02 * w, 0.04 * h, 'blue'),
    ((0.23 * w, 0.5 * h), 0.02 * w, 0.06 * h, 'purple'),
    ((0.73 * w, 0.6 * h), 0.02 * w, 0.07 * h, 'orange'),
    ((0.83 * w, 0.45 * h), 0.015 * w, 0.05 * h, 'yellow'),
    ((0.78 * w, 0.5 * h), 0.05 * w, 0.025 * h, 'green'),
    ((0.2 * w, 0.65 * h), 0.05 * w, 0.02 * h, 'white'),
    ((0.45 * w, 0.8 * h), 0.02 * w, 0.06 * h, 'purple'),
]

for (center, w, h, color) in rectangles:
    rectangle = mpatches.Rectangle((center[0] - 0.5 * w, center[1] - 0.5 * h), w, h, edgecolor = 'black', facecolor = color)
    patches.append(rectangle)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

for patch in patches:
    ax.add_patch(patch)

ax.imshow(background_image)
ax.set_aspect('equal', adjustable = 'box')
ax.axis('off')
plt.savefig('donut.png')
plt.show()
```
